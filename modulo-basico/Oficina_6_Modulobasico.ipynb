```python
import pandas as pd
import numpy as np

dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],
    'Idade': [28, 34, 29, None, 42],
    'Cidade': ['São Paulo', 'Rio de Janeiro', None, 'Curitiba', 'Belo Horizonte'],
    'Vendas': [150, 200, 300, 400, None]
}

df = pd.DataFrame(dados)
print(df)

clientes_maior_30 = df[df['Idade'] > 30]
print("\nClientes com mais de 30 anos:")
print(clientes_maior_30)

media_idade = df['Idade'].mean()
df['Cidade'] = df['Cidade'].fillna('Desconhecido')
print("\nMédia de idade dos clientes:", media_idade)

soma_vendas = df['Vendas'].sum()
print("\nSoma total das vendas:", soma_vendas)
```

```python
# Importando as bibliotecas necessárias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando um conjunto de dados de exemplo (dados aleatórios)
# Você pode substituir isso por seus próprios dados
data = np.random.normal(loc=20, scale=5, size=1000)  # Média de 20 e desvio padrão de 5

# Criando a curva de densidade
sns.kdeplot(data, bw_adjust=0.5)  # bw_adjust ajusta a suavidade da curva

# Adicionando detalhes ao gráfico
plt.title('Curva de Densidade Suavizada')
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.grid()

# Mostrando o gráfico
plt.show()
```

```python
# Importando as bibliotecas necessárias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando um conjunto de dados de exemplo (dados aleatórios)
data = np.random.normal(loc=20, scale=5, size=1000)  # Média de 20 e desvio padrão de 5

# Criando o histograma e a curva de densidade sobrepostos
sns.histplot(data, bins=30, stat='density', kde=True, color='skyblue', edgecolor='black')

# Adicionando detalhes ao gráfico
plt.title('Histograma com Curva de Densidade')
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.grid()

# Mostrando o gráfico
plt.show()
```
