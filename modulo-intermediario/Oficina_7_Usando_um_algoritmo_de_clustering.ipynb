```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


np.random.seed(0)
n_samples = 300

data = {
    'valor_medio_gasto': np.random.randn(n_samples) * 20 + 200,
    'frequencia_compras': np.random.randn(n_samples) * 5 + 15,
    'categorias_produtos': np.random.randn(n_samples) * 2 + 8
}

df = pd.DataFrame(data)


kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[['valor_medio_gasto', 'frequencia_compras', 'categorias_produtos']])

plt.figure(figsize=(10, 6))
plt.scatter(df['valor_medio_gasto'], df['frequencia_compras'], c=df['Cluster'], cmap='viridis', marker='o')
plt.xlabel('Valor Mdio Gasto')
plt.ylabel('Frequncia de Compras')
plt.title('Segmentao de Clientes - K-means Clustering')
plt.colorbar(label='Grupo')
plt.grid()
plt.show()
```

Adequao do Algoritmo K-means:
O K-means  um algoritmo de clustering amplamente utilizado por sua simplicidade e eficcia na identificao de padres em grandes conjuntos de dados. Ele visa dividir os dados em um nmero fixo de grupos (clusters), minimizando a variao dentro de cada grupo. Nesse caso, a segmentao de clientes com base em suas caractersticas de compra pode ser bem capturada pelo K-means, j que essas caractersticas (valor mdio gasto, frequncia de compras e nmero de categorias) so contnuas e podem ser facilmente agrupadas.
As vantagens do K-means: Simplicidade: Fcil de implementar e interpretar.
Escalabilidade: Funciona bem em grandes datasets, o que  relevante para uma loja online com muitos clientes.
Eficiente: O tempo de execuo  relativamente rpido, especialmente se comparado a algoritmos mais complexos.Limitaes do K-means:
Nmero de Clusters: O usurio precisa definir o nmero de clusters a priori, o que pode no refletir a estrutura verdadeira dos dados.
Sensibilidade a Outliers: O algoritmo pode ser afetado por valores extremos, podendo distorcer os resultados.
Formato dos Clusters: Assume que os clusters tm formas esfricas e tamanhos similares, o que pode no ser o caso para todos os tipos de dados.

Os clusters formados pelo algoritmo K-means podem ser interpretados para gerar insights valiosos sobre o comportamento dos clientes da loja online. Por exemplo, um cluster pode representar clientes fiis, que fazem muitas compras com altos gastos mdios, indicando a necessidade de estratgias de reteno, como ofertas exclusivas. Outro cluster pode incluir compradores ocasionais, que tm menor frequncia de compras e valor mdio, sugerindo a possibilidade de estimul-los com promoes especficas. Alm disso, um grupo diversificado, que abrange uma variedade de categorias de produtos, pode indicar interesses variados, permitindo  loja personalizar recomendaes e melhorar a experincia do cliente. Essas anlises ajudam a orientar decises de marketing e a potencializar vendas.
