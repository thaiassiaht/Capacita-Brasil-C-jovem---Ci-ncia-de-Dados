2. O pré-processamento de texto é essencial por várias razões. Primeiramente, textos brutos costumam conter ruídos, como pontuação e palavras comuns (stopwords), que não são úteis para o modelo. Esses elementos podem interferir na análise e na compreensão do texto pelos algoritmos. Além disso, os modelos de machine learning não conseguem entender texto diretamente; eles precisam de representações numéricas para processar as informações de forma eficaz. Essa transformação é crucial para que o modelo consiga realizar suas previsões. O `TfidfVectorizer`, converte os textos em uma matriz de características numéricas. Neste processo, cada palavra é ponderada de acordo com sua importância no documento, utilizando a técnica TF-IDF. Essa abordagem ajuda a destacar as palavras mais relevantes e a minimizar a influência das irrelevantes.
