Como usada na aula, para esse problema também usarei uma CNN, já que são especializadas em processar imagens e é utilizada amplamente pela sua eficiência nesse quesito

Para a preparação das imagens, precisamos redimensioná-las para que tenham o mesmo tamanho. Pesquisando sobre datasets específicos de frutas, percebi que, geralmente, as fotos já seguem um padrão de tamanho. No entanto, para garantir uniformidade e por uma questão de boas práticas de código, é aconselhável redimensioná-las.

Em seguida, normalizamos os valores dos pixels, dividindo-os por 255, pois as cores no RGB variam de 0 a 255. Com isso, os valores ficam entre 0 e 1, o que é útil por diversos motivos, sendo o principal a facilitação do treinamento da IA. Por fim, dividimos os dados em conjuntos de treino, teste e validação. Essa separação é crucial, pois assegura que durante o treinamento da IA, as imagens utilizadas não sejam as mesmas que serão vistas nos testes, o que melhora a generalização do modelo.

Modelo de rede neural


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Cria o modelo sequencial (camadas em sequência)
modelo = Sequential()
# Primeira camada convolucional - detecta bordas simples
modelo.add(Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)))
# Camada de pooling - reduz o tamanho mantendo características importantes
modelo.add(MaxPooling2D((2, 2)))
# Segunda camada convolucional - detecta padrões mais complexos
modelo.add(Conv2D(64, (3, 3), activation='relu'))
modelo.add(MaxPooling2D((2, 2)))
# Terceira camada convolucional - detecta padrões ainda mais complexos
modelo.add(Conv2D(128, (3, 3), activation='relu'))
modelo.add(MaxPooling2D((2, 2)))
# Transforma a matriz 3D em vetor 1D para a rede neural tradicional
modelo.add(Flatten())
# Camada densa (totalmente conectada) para classificação
modelo.add(Dense(128, activation='relu'))
# Camada de saída - 3 neurônios (um para cada classe: maçã, banana, laranja)
modelo.add(Dense(3, activation='softmax'))
```

Compilação e Treinamento do Modelo

```python
# Configura como o modelo vai aprender
modelo.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])  #Acompanha a porcentagem de acertos

# Treina o modelo com os dados
historico = modelo.fit(imagens_treino, rotulos_treino,
                      epochs=10,  #numero de vezes que o modelo vai ver todo o conjunto de treino
                      validation_data=(imagens_validacao, rotulos_validacao))
```
